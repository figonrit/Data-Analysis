```python
##Modificación de valores erroneos
df.loc[df['Product line'] == "Health and beauty" , 'gross margin percentage'] = 0.2
df.loc[df['Product line'] == "Sports and travel" , 'gross margin percentage'] = 0.3
df.loc[df['Product line'] == "Home and lifestyle" , 'gross margin percentage'] = 0.25
df.loc[df['Product line'] == "Food and beverages" , 'gross margin percentage'] = 0.1
df.loc[df['Product line'] == "Fashion accessories" , 'gross margin percentage'] = 0.20
df.loc[df['Product line'] == "Electronic accessories" , 'gross margin percentage'] = 0.4
```

```python
## Calculo nuevas variables
tax = 0.05
df['Unit Cost'] = df['Unit price'] * (1- tax - df['gross margin percentage'])
df['Total'] = df['Unit price'] * df['Quantity']
df['cogs'] = df['Unit Cost'] * df['Quantity']
df['Tax 5%'] = df['Unit price'] * tax
df['Total tax'] = df['Tax 5%'] * df['Quantity']
df['gross income'] = df['Total'] - df['Total tax'] - df['cogs']
```

```python
df = df.reindex(['Invoice ID','Branch','City','Customer type','Gender','Product line','Unit price','Unit Cost','Quantity','Tax 5%','Total tax','Total','cogs','gross margin percentage','gross income','Payment','Rating','Date','Time'], axis=1)
```

```python
df_fechas = df["Date"].str.split("/", expand = True) 
df_fechas.columns = ("Mes","Dia","Año")
df_fechas
```

```python
df = df.join(df_fechas)
```

```python
df = df.reindex(['Invoice ID','Branch','City','Customer type','Gender','Product line','Unit price','Unit Cost','Quantity','Tax 5%','Total tax','Total','cogs','gross margin percentage','gross income','Payment','Rating','Mes','Dia','Año','Time'], axis=1)
```

```python
df2 = pd.DataFrame(df.groupby(["Gender"])["Gender"].count())
df2 = df2.rename(columns={"Gender":"Cantidad"})
df2 = df2.reset_index()

fig = plt.figure(figsize=(8,6))
ax = fig.gca()
sns.barplot(df2["Gender"],df2["Cantidad"], ax=ax, palette="crest")
ax.set_title('Cantidad de Clientes por género', size=17)
ax.set_ylabel("Cantidad", size = 14)
ax.set_xlabel("Gender", size = 14)
for i,rect in zip(df2.index, ax.patches):
    x,y = rect.xy
    ax.text(i-0.12 , y + 250, df2.loc[i]["Cantidad"], color='white', fontweight = 'bold', fontsize = 25)
```

```python
df2 = pd.DataFrame(df.groupby(["Product line"])["Product line"].count())
df2 = df2.rename(columns={"Product line":"Cantidad"})
df2 = df2.reset_index()

fig = plt.figure(figsize=(16,6))
ax = fig.gca()
sns.barplot(df2["Product line"],df2["Cantidad"], ax=ax, palette="crest")
ax.set_title('Cantidad vendida por cada producto', size=17)
ax.set_ylabel("Cantidad", size = 14)
ax.set_xlabel("Product line", size = 14)
for i,rect in zip(df2.index, ax.patches):
    x,y = rect.xy
    ax.text(i-0.15 , y + 85, df2.loc[i]["Cantidad"], color='white', fontweight = 'bold', fontsize = 25)
```

```python
df2 = pd.DataFrame(df.groupby(["Payment"])["Payment"].count())
df2 = df2.rename(columns={"Payment":"Cantidad"})
df2 = df2.reset_index()

fig = plt.figure(figsize=(16,6))
ax = fig.gca()
sns.barplot(df2["Payment"],df2["Cantidad"], ax=ax, palette="crest")
ax.set_title('Cantidad vendida por método de pago', size=17)
ax.set_ylabel("Cantidad", size = 14)
ax.set_xlabel("Payment", size = 14)
for i,rect in zip(df2.index, ax.patches):
    x,y = rect.xy
    ax.text(i-0.1 , y + 175, df2.loc[i]["Cantidad"], color='white', fontweight = 'bold', fontsize = 25)
```

```python
df_pivot = df.pivot_table(index = "Payment", columns = "City", values = "Total", aggfunc={np.sum})
df_pivot
```
